```python
#print odd numbers from 1 to 100 using list comprehension
l=[]
for i in range(1,101):
    l.append(i)
l = [i for i in l if i%2!=0]
print(l)
```

```python
#palindrome function
s = input("Enter a string")
i = 0
while(i<len(s)):
    s1 = s[::-1]
    if(s1 == s):
        print("Given string is a palindrome")
        break
    else:
        print("Given string isn't a palindrome")
        break
    i = i+1
```
